```python
# Model Implementation

class LinearRegression:
    def __init__(self, W_init='Norm', method="batch", loss_fn='norm', lam=0.05,
    max_iter=10000, tol=0.01, alpha=0.00000002, batch_size=100):
        self.method = method
        self.max_iter = max_iter
        self.tol = tol
        self.alpha = alpha
        self.batch_size = batch_size
        self.W_init = W_init
        self.loss_fn = loss_fn
        self.lam = lam


    def fit(self, X_train, y_train):
        m, n = X_train.shape
        self.iter_stop = 0
        if self.W_init=='norm':
            self.theta = np.random.normal(0, 1, n)
        elif self.W_init=='rand':
            self.theta = np.random.uniform(0, 1, n)
        elif self.W_init=='zero':
            self.theta = np.zeros(n)
        else:
            raise ValueError("ValueError: W init is not match ('norm', 'rand', 'zero')")
        
        list_of_used_ix = []
        self.losses = []
        loss_old = 10000
        for i in range(self.max_iter):
            if self.method == "batch":
                self.X_train = X_train
                self.y_train = y_train
            elif self.method == "sto":
                idx = np.random.randint(0, m)
                while idx in list_of_used_ix:
                    idx = np.random.randint(m)
                self.X_train = X_train[idx, :].reshape(1, -1)
                self.y_train = y_train[idx]
                list_of_used_ix.append(idx)
                if len(list_of_used_ix) == m: list_of_used_ix = []
            elif self.method == "mini":
                idx = np.random.randint(0, m-self.batch_size)
                self.X_train = X_train[idx:idx+self.batch_size]
                self.y_train = y_train[idx:idx+self.batch_size]
            else:
                raise ValueError("ValueError: Method is not match ('batch', 'sto', 'mini')")
            
            yhat = self.hx(self.X_train, self.theta)
            error = yhat - self.y_train
            grad = self.gradient(self.X_train, error)
            self.theta = self.theta - self.alpha * grad

            loss_new = self.Loss(yhat, self.y_train)
            diff = abs(loss_new - loss_old)

            self.losses.append(loss_new)

            self.iter_stop = i+1
            if diff < self.tol:
                break
            else:
                loss_old = loss_new

    def evalute(self, X_test, y_test):
        yhat_test = self.hx(X_test, self.theta)
        mse = (((yhat_test - y_test)**2).sum())/len(y_test)
        return mse

    def iter_stop(self):
        return self.iter_stop

    def time_taken(self):
        return self.time_take

    def hx(self, X, theta):
        return X @ theta

    def Loss(self, yhat, y):
        if self.loss_fn == 'norm':
            loss = 0.5*(((yhat - y)**2).sum())
        elif self.loss_fn == 'ridge':
            #print((self.theta**2).sum)
            loss = 0.5*(((yhat - y)**2).sum())+(self.lam*(self.theta**2).sum())
        return loss

    def gradient(self, X, error):
        if self.loss_fn == 'norm':
            grad = X.T @ error
        elif self.loss_fn == 'ridge':
            grad = (X.T @ error) + (self.lam*self.theta)
        else:
            raise ValueError("ValueError: loss fn is not match ('norm', 'ridge')")
        return grad

    def predict(self, X_test):
        return self.hx(X_test, self.theta)
```

```python
def CV(X_train_val, y_train_val, model, cv=3):
    # Split train data and validation data
    m, n = X_train_val.shape
    idx = list(range(m))
    idx_List = []
    for i in range(cv):
        idx_List.append(idx[i*int(m/cv):(i+1)*int(m/cv)])
    # Predict and find accuracy
    acc = 0
    for idx in idx_List:
        X_val = X_train_val[idx]
        y_val = y_train_val[idx]
        X_train = np.delete(X_train_val,idx, axis=0)
        y_train = np.delete(y_train_val,idx, axis=0)
        model.fit(X_train, y_train)
        acc += model.evalute(X_val, y_val)
    return acc/cv
```

```python
model_list = [LinearRegression(W_init='zero', method="mini", loss_fn='norm'),
LinearRegression(W_init='zero', method="mini", loss_fn='ridge', lam=0.05),
LinearRegression(W_init='zero', method="sto", loss_fn='norm'),
LinearRegression(W_init='zero', method="sto", loss_fn='ridge', lam=0.05)]

for model in model_list:
    acc = CV(X_train_norm, y_train, model, cv=3)
    print("Accuracy:",acc)
```
